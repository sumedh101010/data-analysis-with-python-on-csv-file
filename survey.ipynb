```python
learn_methods=data['LearnCode'].str.split(";").explode() # it gives u a detailed information thats why explode is use 
```

```python
learn_methods.value_counts()
```

```python
learn_methods.value_counts(normalize=True)*100
```

```python
method_counts=learn_methods.value_counts()
method_counts
```

```python
method_percentages=(method_counts/89184)*100
method_percentages
```

```python
from matplotlib import pyplot as plt 

method_percentages.plot(kind='barh')  # barh means bar graph in horizontal fashion
plt.title("Learning methods")
plt.xlabel('Percentage')
plt.ylabel('Learn codes')
for index,value in enumerate(method_percentages):
     plt.text(value,index,f'{value:.2f}%')  # 2f upto 2 decimal places
plt.show()
```
